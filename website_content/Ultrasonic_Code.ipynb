```python
import time
from Raspbot_Lib import Raspbot
from ipywidgets import interact
import ipywidgets as widgets
import sys
sys.path.append('/home/pi/software/oled_yahboom/')
from yahboom_oled import *
```

```python
bot = Raspbot()

#Creating a OLED object (organic light-emitting diode) for display
oled = Yahboom_OLED(debug=False)
```

```python
def Display_Data(distance):
    disFormat = "{:>10} mm".format(distance)
    line = "Ultrasonic distance: " + disFormat
    oled.add_line(line, 1)
    oled.refresh()

    if (distance<100):
        oled.add_line("Too Close", 3)
    elif(distance<300):
        oled.add_line("Close", 3)
    else:
        oled.add_line("Safe", 3)
```

```python

bot.Ctrl_Ulatist_Switch(1)
time.sleep(0.03) #allow time for ultrasonic sensor to measure
oled.init_oled_process()
try:
    while(1):
        diss_H =bot.read_data_array(0x1b,1)[0]
        diss_L =bot.read_data_array(0x1a,1)[0]
        distance = diss_H << 8 | diss_L #combine the both data into one
        Display_Data(distance)
        time.sleep(0.05)
except KeyboardInterrupt:
    print("Done\n")
finally:
    bot.Ctrl_Ulatist_Switch(0)
    os.system("python3 /home/pi/software/oled_yahboom/yahboom_oled.py &")
    print("Cleaned up")
```
