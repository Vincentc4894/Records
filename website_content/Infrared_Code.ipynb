```python
import sys
sys.path.append('/home/pi/project_demo/lib')
from McLumk_Wheel_Sports import *
import time
```

```python
speed =20
bot = Raspbot()
```

```python
def Identify_Angle (x1, x2, x3, x4):
    if (x1 ==0 and x2 ==0 and x3 ==0 and x4 ==0):
        print ("Black")
        return "Black"
    elif (x4 == 0 and (x1 ==0 or x2 ==0)):
        print("Right Shift")
        return "Right Shift"
    elif(x1 == 0 and (x3 ==0 or x4 ==0)):
        print("Left Shift")
        return "Left Shift"
    elif(x1 == 0):
        print("Outer Left")
        return "Outer Left"
    elif(x4 ==0):
        print ("Outer Right")
        return "Outer Right"
    elif (x2 == 0 and x3 == 1):
        print("Adjust Left")
        return "Adjust Left"
    elif (x2 == 1 and x3 == 0):
        print("Adjust Right")
        return "Adjust Right"
    elif (x2 == 0 and x3 == 0):
        print("All Black")
        return "All Black"
    else:
        print("Stop")
        return "Stop"
```

```python
def Angle_Based_Movement(angle):
    if (angle == "Black" or angle == "All Black"):
        move_forward(speed)
    elif (angle == "Right Shift" or angle == "Outer Right" or angle == "Adjust Right"):
        rotate_right(speed)
    elif (angle == "Left Shift" or angle == "Outer Left" or angle == "Adjust Left" ):
        rotate_left(speed)
    return
```

```python
while(True):
    #reading form I2C
    track_data = bot.read_data_array(0x0a, 1)

    track = int(track_data[0])

    #there are 4 sensors. assign each bit to a variable
    l2 = (track >> 3) & 0x01 # lower left
    l1 = (track >> 2) & 0x01 # upper left
    r1 = (track >> 1) & 0x01 # upper right
    r2 = track & 0x01 #lower right

    #using if statement to indentify next course of action
    angle = Identify_Angle(l1,l2,r1,r2)

    #should only happen if it's all white meaning path ended
    if(angle == "Stop"):
        break

    #adjustment at need
    Angle_Based_Movement(angle)

    #pause before next action
    time.sleep (0.05)
```
