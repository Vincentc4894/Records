```python
import sys
import time
sys.path.append('/home/pi/project_demo/lib')
from McLumk_Wheel_Sports import *
```

```python
speed  = 50 #you can change the speed at which it moves it moves
duration = 2 # number of second between each code.
```

```python
def sideways(speed, duration):
    #going right
    Ctrl_Muto(1, speed)
    Ctrl_Muto(2, -speed)
    Ctrl_Muto(3, -speed)
    Ctrl_Muto(4, speed)
    time.sleep(duration)
    
    #pause
    Ctrl_Muto(1, 0)
    Ctrl_Muto(2, 0)
    Ctrl_Muto(3, 0)
    Ctrl_Muto(4, 0)
    time.sleep(1)

    #going left to the starting position
    Ctrl_Muto(1, -speed)
    Ctrl_Muto(2, speed)
    Ctrl_Muto(3, speed)
    Ctrl_Muto(4, -speed)
    time.sleep(duration)
    
    #return to rest
    Ctrl_Muto(1, 0)
    Ctrl_Muto(2, 0)
    Ctrl_Muto(3, 0)
    Ctrl_Muto(4, 0)
```

```python
#Let first test to see which motor Mx is attach to x={1,2,3,4}
Ctrl_Muto(1, speed)
time.sleep(duration)
Ctrl_Muto(1, 0)

Ctrl_Muto(2, speed)
time.sleep(duration)
Ctrl_Muto(2, 0)

Ctrl_Muto(3, speed)
time.sleep(duration)
Ctrl_Muto(3, 0)

Ctrl_Muto(4, speed)
time.sleep(duration)
Ctrl_Muto(4, 0)

#You can also observe the same effect happening using Ctrl_Car()
Ctrl_Car(4, 0, speed)
time.sleep(duration)
Ctrl_Car(4, 0, 0)
```

```python
#Now let uses all the motors to in sychronation to made it go right then left back to starting square
sideways(speed, duration)
time.sleep(duration)
```
