```python
import cv2
import ipywidgets.widgets as widgets
import threading
import time

#display() may have been included already, if not uncomment this library.
#from IPython.display import display
```

```python
#converting information into jpeg
def bgr8_to_jpeg(value):
    return bytes(cv2.imencode('.jpg', value)[1])
```

```python
#connecting to camera
image = cv2.VideoCapture(0)

#your welcome to set the camera setting as needed
```

```python
image_widget = widgets.Image(format='jpeg')
display(image_widget)

#displaying a black image temperary 
ret, frame = image.read()
black = frame*0
image_widget.value = bgr8_to_jpeg(black)
```

```python
#capturing one frame at a time
try:
    while ret :
        image_widget.value = bgr8_to_jpeg(frame)
        #slight pause between each frame
        time.sleep(0.03)
        #updating value
        ret, frame = image.read()
except KeyboardInterrupt:
    print("Forced Closed")
```

```python
#Make sure to run this, or future code requiring camera will get stuck
#the camera will remain connect to this code and reject other code if you fail to run this fully.
image.release()
cv2.destroyAllWindows()
```
